```python
from tkinter import *
from tkinter import messagebox
import random

#Herstellung des Spielfenster/Interface
window = Tk()
window.geometry("455x690")
window.title("Tic-Tac-Toe")

#Spieler und random choice für Startenderspieler
players = ["x","o"]
player = random.choice(players)

#Darstellen des Spielfeldes mit den 9 Buttons
buttons = [[0,0,0],
           [0,0,0],
           [0,0,0]]

#Text für die Interface
label = Label(text=player + "'s turn", font=('consolas',40))
label.pack(side="top")




#Definition des Spielneustarts, alle Buttons kehren zum anfangszustand, der count wird reseted und die Buttons werden wieder aktiviert
def new_game():
    global player, count

    player = random.choice(players)

    label.config(text=player+"'s turn")
    
    buttons[0][0].config(text="", bg="#F0F0F0")
    buttons[0][1].config(text="", bg="#F0F0F0")
    buttons[0][2].config(text="", bg="#F0F0F0")
    
    buttons[1][0].config(text="", bg="#F0F0F0")
    buttons[1][1].config(text="", bg="#F0F0F0")
    buttons[1][2].config(text="", bg="#F0F0F0")
    
    buttons[2][0].config(text="", bg="#F0F0F0")
    buttons[2][1].config(text="", bg="#F0F0F0")
    buttons[2][2].config(text="", bg="#F0F0F0")
    
    count=0
    
    enable_buttons()
    
#Definition einer Funktion um das SPiel zu verlassen    
def exit_game():
    window.destroy()

#Reset Button für die Interface
reset_button = Button(text="restart", font=('consolas',20), bg="plum1", command=new_game)
reset_button.pack(side="top")

#Exit button um das Spiel zu schliessen
exit_button = Button(text="exit", font=('consolas',20), bg="plum1", command=exit_game)
exit_button.place(x=290, y=68)

#Herstellung eines Frames und einfügen ind dem Hauptfenster
frame = Frame(window)
frame.pack()


#Herstellung eines 3x3-Raster von Buttons im frame
for row in range(3):
    for column in range(3):
        buttons[row][column] = Button(frame, text="",font=('consolas',40), width=5, height=2,relief="groove",
                                      command= lambda row=row, column=column: next_turn(row,column))
        buttons[row][column].grid(row=row,column=column)

#Defintion um die Knöpfe auszuschalten 
def disable_buttons():
    buttons[0][0]['state'] = DISABLED
    buttons[0][1]['state'] = DISABLED
    buttons[0][2]['state'] = DISABLED
    buttons[1][0]['state'] = DISABLED
    buttons[1][1]['state'] = DISABLED
    buttons[1][2]['state'] = DISABLED
    buttons[2][0]['state'] = DISABLED
    buttons[2][1]['state'] = DISABLED
    buttons[2][2]['state'] = DISABLED

#Defintion um die Knöpfe anzuschalten    
def enable_buttons():
    buttons[0][0]['state'] = NORMAL
    buttons[0][1]['state'] = NORMAL
    buttons[0][2]['state'] = NORMAL
    buttons[1][0]['state'] = NORMAL
    buttons[1][1]['state'] = NORMAL
    buttons[1][2]['state'] = NORMAL
    buttons[2][0]['state'] = NORMAL
    buttons[2][1]['state'] = NORMAL
    buttons[2][2]['state'] = NORMAL



count=0

Tie="Tie"
 
#Defintion die den Spielausgang analysiert    
def check_winner():
    global count
    

    if buttons[0][0]['text'] == buttons[0][1]['text'] == buttons[0][2]['text'] != '':
        buttons[0][0].config(bg="blue")
        buttons[0][1].config(bg="blue")
        buttons[0][2].config(bg="blue")
        return True

    elif buttons[1][0]['text'] == buttons[1][1]['text'] == buttons[1][2]['text'] != '':
        buttons[1][0].config(bg="blue")
        buttons[1][1].config(bg="blue")
        buttons[1][2].config(bg="blue")
        return True

    elif buttons[2][0]['text'] == buttons[2][1]['text'] == buttons[2][2]['text'] != '':
        buttons[2][0].config(bg="blue")
        buttons[2][1].config(bg="blue")
        buttons[2][2].config(bg="blue")
        return True

    elif buttons[0][0]['text'] == buttons[1][0]['text'] == buttons[2][0]['text'] != '':
        buttons[0][0].config(bg="blue")
        buttons[1][0].config(bg="blue")
        buttons[2][0].config(bg="blue")
        return True

    elif buttons[0][1]['text'] == buttons[1][1]['text'] == buttons[2][1]['text'] != '':
        buttons[0][1].config(bg="blue")
        buttons[1][1].config(bg="blue")
        buttons[2][1].config(bg="blue")
        return True

    elif buttons[0][2]['text'] == buttons[1][2]['text'] == buttons[2][2]['text'] != '':
        buttons[0][2].config(bg="blue")
        buttons[1][2].config(bg="blue")
        buttons[2][2].config(bg="blue")
        return True

    elif buttons[2][0]['text'] == buttons[1][1]['text'] == buttons[0][2]['text'] != '':
        buttons[2][0].config(bg="blue")
        buttons[1][1].config(bg="blue")
        buttons[0][2].config(bg="blue")
        return True

    elif buttons[0][0]['text'] == buttons[1][1]['text'] == buttons[2][2]['text'] != '':
        buttons[0][0].config(bg="blue")
        buttons[1][1].config(bg="blue")
        buttons[2][2].config(bg="blue")
        return True

    
    
    if count == 9: 
        return "Tie"
    
    else:
        return False
        
#Defintion next_turn führt einen Spielzug durch und überprüft ob jemand das Spiel gewonnen hat
def next_turn(row, column):
    global player, winner, count 
    if buttons[row][column]['text'] == "":
        buttons[row][column]['text'] = player
        count += 1
        if check_winner() is False:
            player = players[1] if player == players[0] else players[0]
            label.config(text=player + "'s turn")
        elif check_winner() is True:
            label.config(text= "Hurray "  +  player + " wins")
            disable_buttons()
        elif check_winner() == Tie:
            label.config(text="It's a Tie")
            disable_buttons()
    else:
        messagebox.showerror(title=None, message='This square is already taken. Please choose another one')

    
    


window.mainloop()
```
